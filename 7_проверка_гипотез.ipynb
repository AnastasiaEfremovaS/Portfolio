–£ –≤—Å–µ—Ö —Ç—Ä–µ—Ö —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω–æ –Ω–µ–±–æ–ª—å—à–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–æ–≤–µ—Ä–∫–∏ (Efforts), –Ω–æ –ø—Ä–∏ —ç—Ç–æ–º –≤–ø–µ—á–∞—Ç–ª—è—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ –ø–æ –≤–ª–∏—è–Ω–∏—é –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è (Impact) –∏ —Å—Ç–µ–ø–µ–Ω–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –≤ —ç—Ç–æ–º –≤–ª–∏—è–Ω–∏–∏ (Confidence).

1. —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö –Ω–µ—Ç
2. –ø—Ä–æ–ø—É—Å–∫–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç, –Ω–æ —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –Ω–µ –≤–µ–∑–¥–µ –ø–æ–¥—Ö–æ–¥—è—â–∏–π - –¥–∞—Ç—ã –≤ –æ–±–æ–∏—Ö –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞—Ö –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ —Ç–∏–ø—É datetime
3. —Å–∫—Ä—ã—Ç—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –±—ã–ª–∏ –≤–æ–∑–º–æ–∂–Ω—ã —Ç–æ–ª—å–∫–æ –≤ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è—Ö –≥—Ä—É–ø–ø, –Ω–æ –æ–Ω–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç - –≤ –¥–∞–Ω–Ω—ã—Ö —Ç–æ–ª—å–∫–æ –≥—Ä—É–ø–ø—ã –ê –∏ –í.

–ü—Ä–∏–≤–µ–¥–µ–º –¥–∞—Ç—ã –∫ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–º—É —Ñ–æ—Ä–º–∞—Ç—É

```python
#Change date format to datetime
orders['date'] = pd.to_datetime(orders['date'],format='%Y-%m-%d')
visitors['date'] = pd.to_datetime(visitors['date'],format='%Y-%m-%d')
#Check new date types
print(orders['date'].dtype)
print(visitors['date'].dtype)
```

–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏–º –¥–∞—Ç—ã, –≤ —Ä–∞–º–∫–∞—Ö –∫–æ—Ç–æ—Ä—ã—Ö –∑–∞–∫–∞–∑—á–∏–∫–∏ –ø–æ—Å–µ—â–∞–ª–∏ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω –∏ —Å–æ–≤–µ—Ä—à–∞–ª–∏ –ø–æ–∫—É–ø–∫–∏, —á—Ç–æ–±—ã —É–±–µ–¥–∏—Ç—å—Å—è —á—Ç–æ –æ–Ω–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –¥–∞—Ç–∞–º —Ç–µ—Å—Ç–∞.

```python
#check min and max date of visits
print('–ù–∞—á–∞–ª—å–Ω–∞—è –¥–∞—Ç–∞ –ø–æ—Å–µ—â–µ–Ω–∏—è –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω–∞:')
print(visitors['date'].min())
print('')
print('–ö–æ–Ω–µ—á–Ω–∞—è –¥–∞—Ç–∞ –ø–æ—Å–µ—â–µ–Ω–∏—è –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-–º–∞–≥–∞–∑–∏–Ω–∞:')
print(visitors['date'].max())
print('')
#check min and max date of orders
print('–î–∞—Ç–∞ –ø–µ—Ä–≤–æ–≥–æ –∑–∞–∫–∞–∑–∞:')
print(orders['date'].min())
print('')
print('–î–∞—Ç–∞ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –∑–∞–∫–∞–∑–∞:')
print(orders['date'].max())
```

–ù–∞—á–∞–ª—å–Ω–∞—è –¥–∞—Ç–∞ - –∫–∞–∫ –≤–∏–∑–∏—Ç–æ–≤, —Ç–∞–∫ –∏ –∑–∞–∫–∞–∑–æ–≤ - 1 –∞–≤–≥—É—Å—Ç–∞ 2019 –≥–æ–¥–∞, –∫–æ–Ω–µ—á–Ω–∞—è –¥–∞—Ç–∞ - 31 –∞–≤–≥—É—Å—Ç–∞ 2019 –≥–æ–¥–∞.
–¢–µ–ø–µ—Ä—å —É –Ω–∞—Å –µ—Å—Ç—å –ø–æ–Ω–∏–º–∞–Ω–∏–µ —Å—Ä–æ–∫–æ–≤ —Ç–µ—Å—Ç–∞ - –æ–¥–∏–Ω –º–µ—Å—è—Ü, –∞–≤–≥—É—Å—Ç 2019 –≥–æ–¥–∞.

–¢–∞–∫–∂–µ –ø—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ –≤ –≥—Ä—É–ø–ø–∞—Ö –ê –∏ –í —É—á–∞—Å—Ç–≤—É—é—Ç —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏, —Ç.–∫. –¥–ª—è —Ç–µ—Å—Ç–∞ –≤–∞–∂–Ω–æ —á—Ç–æ–±—ã –æ–Ω–∏ –±—ã–ª–∏ —á–µ—Ç–∫–æ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã: –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏, —É—á–∞—Å—Ç–≤—É—é—â–∏–µ –≤ –æ–±–µ–∏—Ö –≥—Ä—É–ø–ø–∞—Ö, –Ω–µ –ø–æ–∑–≤–æ–ª—è—Ç –ø—Ä–æ–≤–µ—Ä—Å—Ç–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏. 

–°–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–∫–∏ –∏–∑ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –ê –∏ –í, –∑–∞—Ç–µ–º —Å–æ–±–µ—Ä–µ–º –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫ –∏–∑ –ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è ID –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –µ—Å–ª–∏ —Ç–∞–∫–∏–µ –µ—Å—Ç—å.

```python
#let's create two lists with unique users' ID
group_b = orders[orders['group'] == 'B']['visitorId'].unique()
group_a = orders[orders['group'] == 'A']['visitorId'].unique()
#now we use cicle to find any non-unique users in both groups
dup_visitors = []
for visitor in group_b:
    if visitor in group_a:
        dup_visitors.append(visitor)
#count number of non-unique users:
print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ–≤—Ç–æ—Ä—è—é—â–∏—Ö—Å—è –≤ –≥—Ä—É–ø–ø–∞—Ö –ê –∏ –í')
print(len(dup_visitors))
```

–ú–æ–∂–Ω–æ —É–¥–∞–ª–∏—Ç—å –≤—Å–µ –∑–∞–ø–∏—Å–∏ —Å —ç—Ç–∏–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ –∏–∑ –≥—Ä—É–ø–ø—ã –í –≤ —Ç–∞–±–ª–∏—Ü–µ orders, –Ω–æ –º—ã –Ω–µ –º–æ–∂–µ–º —É–¥–∞–ª–∏—Ç—å —ç—Ç–∏—Ö –∂–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ —Ç–∞–±–ª–∏—Ü—ã visitors - —Ç.–∫. —Ç–∞–º –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö –æ–± –∏—Ö ID.

–û—Ü–µ–Ω–∏–º, –∫–∞–∫–æ–π –ø—Ä–æ—Ü–µ–Ω—Ç —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç 58 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ —á–∏—Å–ª–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã –í.

```python
#show the proportion of 58 in unique users of gourp B
print('–ü—Ä–æ—Ü–µ–Ω—Ç –Ω–µ—É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å—Ä–µ–¥–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≥—Ä—É–ø–ø—ã B')
print(58/ orders[orders['group'] == 'B']['visitorId'].nunique())
```

–í —Ü–µ–ª–æ–º —Ç–∞–∫–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤—Å–µ–≥–¥–∞ —Å—Ç–æ–∏—Ç —É–¥–∞–ª—è—Ç—å, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∏ –º–æ–≥—É—Ç –¥–∞–≤–∞—Ç—å —Å–º–µ—â–µ–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏ —Ü–µ–ª–µ–≤–æ–≥–æ —Å–æ–±—ã—Ç–∏—è: –µ—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å, —É—á–∞—Å—Ç–≤—É—é—â–∏–π –≤ –æ–±–æ–∏—Ö –≥—Ä—É–ø–ø–∞—Ö, —Å–æ–≤–µ—Ä—à–∏–ª —Ü–µ–ª–µ–≤–æ–µ —Å–æ–±—ã—Ç–∏–µ —Ç–µ—Å—Ç–∞, –º—ã –Ω–µ —Å–º–æ–∂–µ–º –¥–æ—Å—Ç–æ–≤–µ—Ä–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å, –ø—Ä–æ–∏–∑–æ—à–ª–æ —ç—Ç–æ –ø–æ–¥ –≤–ª–∏—è–Ω–∏–µ–º –Ω–æ–≤–æ–≤–≤–µ–¥–µ–Ω–∏—è, –∏–ª–∏ –∂–µ —ç—Ç–æ –µ–≥–æ –æ–±—ã—á–Ω–æ–µ –ø–æ–≤–µ–¥–µ–Ω–∏–µ. 
    
–ù–æ –≤ —Ä–∞–º–∫–∞—Ö —É—á–µ–±–Ω–æ–≥–æ –ø—Ä–æ–µ–∫—Ç–∞ –º–æ–∂–Ω–æ –æ—Å—Ç–∞–≤–∏—Ç—å:)

–ü—Ä–∏—Å—Ç—É–ø–∏–º –∫ –∞–Ω–∞–ª–∏–∑—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –ê/–í —Ç–µ—Å—Ç–∞.

–î–ª—è –Ω–∞—á–∞–ª–∞, –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º.

```python
# create df with unique dates and groups
datesGroups = orders[['date','group']].drop_duplicates()

# obtain aggregated orders by days 
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'transactionId' : 'nunique', 'visitorId' : 'nunique', 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])

# obtain aggregated number of visitors by days 
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])

# merge cumulative data into one table with clear columns' names
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
```

–¢–µ–ø–µ—Ä—å —Å–æ–±–µ—Ä–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –∏ –≤—ã—á–∏—Å–ª–∏–º p-value –∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ –≥—Ä—É–ø–ø—ã –í –ø–µ—Ä–µ–¥ –≥—Ä—É–ø–ø–æ–π –ê.

```python
# group orders by users in groups A and B
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders'] 
#create samples of users by groups with number of orders
sampleA = pd.concat([ordersByUsersA['orders'],pd.Series(0, index=np.arange(visitors[visitors['group'] == 'A']['visitors'].\
                                                                           sum() - len(ordersByUsersA['orders'])),\
                                                        name='orders')],axis=0)

sampleB = pd.concat([ordersByUsersB['orders'],pd.Series(0, index=np.arange(visitors[visitors['group'] == 'B']['visitors'].\
                                                                           sum() - len(ordersByUsersB['orders'])), \
                                                        name='orders')],axis=0) 
print('P-value —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç:')
print("{0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ –≥—Ä—É–ø–ø—ã –í –ø–µ—Ä–µ–¥ –≥—Ä—É–ø–ø–æ–π –ê:')
print("{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

```python
#create samples with filtered data
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(0, index=np.arange(visitors[visitors['group'] == 'A']['visitors'].sum() - len(ordersByUsersA['orders'])),
            name='orders',),],axis=0,)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(0,index=np.arange(visitors[visitors['group'] == 'B']['visitors'].sum() - len(ordersByUsersB['orders'])),
            name='orders',),],axis=0,) 

print('P-value —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç:')
print('{0:.3f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ –≥—Ä—É–ø–ø—ã B –ø–µ—Ä–µ–¥ –≥—Ä—É–ø–ø–æ–π –ê:')
print('{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1)) 
```

–í—ã–≤–æ–¥: –ü–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º —Ç–∞–∫–∂–µ –º–æ–∂–Ω–æ –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É - –æ—Ç–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º —á–∏—Å–ª–µ –∑–∞–∫–∞–∑–æ–≤ –¥–ª—è –≥—Ä—É–ø–ø A –∏ B –¥–æ—Å—Ç–∏–≥–ª–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, —Ç.–∫. p-value —Ä–∞–≤–Ω–æ 0.018 - –º–µ–Ω—å—à–µ 0,05.

–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –≤—ã–∏–≥—Ä—ã—à –≥—Ä—É–ø–ø—ã B —Ä–∞–≤–µ–Ω 14.5% ‚Äî –¥–∞–∂–µ –≤—ã—à–µ —á–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ "—Å—ã—Ä—ã—Ö" –¥–∞–Ω–Ω—ã—Ö.

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> –û—Ü–µ–Ω–∏–ª–∏ –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –Ω–∞ –æ—á–∏—â–µ–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö - –æ—Ç–ª–∏—á–Ω–æ!

–¢–µ–ø–µ—Ä—å –ø–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º.

```python
# calculate and show difference between filtered average cheques for both groups
print('P-value —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç:')
print(
    '{0:.3f}'.format(
        stats.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )[1]
    )
)
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ –≥—Ä—É–ø–ø—ã B –ø–µ—Ä–µ–¥ –≥—Ä—É–ø–ø–æ–π –ê:')
print(
    "{0:.3f}".format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        - 1
    )
) 
```
